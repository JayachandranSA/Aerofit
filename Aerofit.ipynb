#**Q2. OUTLIER DETECTION**

```python
fig = plt.figure(figsize=(12,8)).suptitle("Detecting Outliers",fontsize=20)

plt.subplot(2,3,1)
sns.boxplot(data =AerofitDb, x ="Age", orient = "h",color ="red")
plt.subplot(2,3,2)
sns.boxplot(data =AerofitDb, x ="Education", orient = "h",color ="blue")
plt.subplot(2,3,3)
sns.boxplot(data =AerofitDb, x="Usage", orient = "h",color ="green")
plt.subplot(2,3,4)
sns.boxplot(data =AerofitDb, x="Fitness", orient = "h",color ='yellow')
plt.subplot(2,3,5)
sns.boxplot(data =AerofitDb, x="Income", orient = "h",color ="pink")
plt.subplot(2,3,6)
sns.boxplot(data =AerofitDb, x="Miles", orient = "h", color ="magenta")
plt.show()
```

**AGE**

```python
Age_25 = np.percentile(AerofitDb["Age"], 25)
print(f"25 Percentile :{Age_25}")
Age_75 = np.percentile(AerofitDb["Age"], 75)
print(f"75 percentile :{Age_75}")
# Inter Quartile Range
Age_IQR = Age_75 - Age_25
print(f"Inter Quartile Range of Age:{Age_IQR}")
upper_limit = Age_75 + 1.5*Age_IQR
print(f"upper_limit:{upper_limit}")
outliers_Age= AerofitDb[AerofitDb["Age"]>upper_limit]
len(outliers_Age)
per_outliers_Age =  (len(outliers_Age)/len(AerofitDb))*100
print(f"Percentage Of Outlier for Age :{per_outliers_Age}")
```

**EDUCATION**

```python
Education_25 = np.percentile(AerofitDb["Education"], 25)
print(f"25 Percentile :{Education_25}")
Education_75 = np.percentile(AerofitDb["Education"], 75)
print(f"75 percentile :{Education_75}")
# Inter Quartile Range
Education_IQR = Education_75 - Education_25
print(f"Inter Quartile Range of Education:{Education_IQR}")
upper_limit = Education_75 + 1.5*Education_IQR
print(f"upper_limit:{upper_limit}")
outliers_Education= AerofitDb[AerofitDb["Education"]>upper_limit]
len(outliers_Education)
per_outliers_Education =  (len(outliers_Education)/len(AerofitDb))*100
print(f"Percentage Of Outlier for Education :{per_outliers_Education}")

```

**USAGE**

```python
Usage_25 = np.percentile(AerofitDb["Usage"], 25)
print(f"25 Percentile :{Usage_25}")
Usage_75 = np.percentile(AerofitDb["Usage"], 75)
print(f"75 percentile :{Usage_75}")
# Inter Quartile Range
Usage_IQR = Usage_75 - Usage_25
print(f"Inter Quartile Range of Usage:{Usage_IQR}")
upper_limit = Usage_75 + 1.5*Usage_IQR
print(f"upper_limit:{upper_limit}")
outliers_Usage= AerofitDb[AerofitDb["Usage"]>upper_limit]
len(outliers_Usage)
per_outliers_Usage =  (len(outliers_Usage)/len(AerofitDb))*100
print(f"Percentage Of Outlier for Usage :{per_outliers_Usage}")
```

**FITNESS**

```python
Fitness_25 = np.percentile(AerofitDb["Fitness"], 25)
print(f"25 Percentile :{Fitness_25}")
Fitness_75 = np.percentile(AerofitDb["Fitness"], 75)
print(f"75 percentile :{Fitness_75}")
# Inter Quartile Range
Fitness_IQR = Fitness_75 - Fitness_25
print(f"Inter Quartile Range of Fitness:{Fitness_IQR}")
upper_limit = Fitness_75 + 1.5*Fitness_IQR
print(f"upper_limit:{upper_limit}")
outliers_Fitness= AerofitDb[AerofitDb["Fitness"]>upper_limit]
len(outliers_Fitness)
per_outliers_Fitness =  (len(outliers_Fitness)/len(AerofitDb))*100
print(f"Percentage Of Outlier for Fitness :{per_outliers_Fitness}")
```

**INCOME**

```python
Income_25 = np.percentile(AerofitDb["Income"], 25)
print(f"25 Percentile :{Income_25}")
Income_75 = np.percentile(AerofitDb["Income"], 75)
print(f"75 percentile :{Income_75}")
# Inter Quartile Range
Income_IQR = Income_75 - Income_25
print(f"Inter Quartile Range of Income:{Income_IQR}")
upper_limit = Income_75 + 1.5*Income_IQR
print(f"upper_limit:{upper_limit}")
outliers_Income= AerofitDb[AerofitDb["Income"]>upper_limit]
len(outliers_Income)
per_outliers_Income =  (len(outliers_Income)/len(AerofitDb))*100
print(f"Percentage Of Outlier for Income :{per_outliers_Income}")

```

**MILES**

```python
Miles_25 = np.percentile(AerofitDb["Miles"], 25)
print(f"25 Percentile :{Miles_25}")
Miles_75 = np.percentile(AerofitDb["Miles"], 75)
print(f"75 percentile :{Miles_75}")
# Inter Quartile Range
Miles_IQR = Miles_75 - Miles_25
print(f"Inter Quartile Range of Miles:{Miles_IQR}")
upper_limit = Miles_75 + 1.5*Miles_IQR
print(f"upper_limit:{upper_limit}")
outliers_Miles= AerofitDb[AerofitDb["Miles"]>upper_limit]
len(outliers_Miles)
per_outliers_Miles =  (len(outliers_Miles)/len(AerofitDb))*100
print(f"Percentage Of Outlier for Miles :{per_outliers_Miles}")

```
